```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('./green_tripdata_2024-01.parquet')
df.head()
```

```python
df = df[df.trip_type==2] #Dispatch rides (booked via app)
```

```python
df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime #Target variable since we want to predict trip duration
df['duration']= df.duration.apply(lambda td: td.total_seconds()/60) #Converting the time duration into minutes 
df.head()
```
